```python
from PIL import Image
import numpy as np

path = "Instagram_icon.png"
img = Image.open(path)

img_arr = np.array(img)

print(img_arr.shape)
```

```python
#   
uniq = np.unique(img_arr[:,:,0:-1].reshape(5001*5001, 3), axis=0)
```

```python
uniq.shape
```

```python
def rgb_to_hex(rgb):
    r, g, b = rgb
    r, g, b = int(r), int(g), int(b)
    return '#' + hex(r)[2:].zfill(2) + hex(g)[2:].zfill(2) + hex(b)[2:].zfill(2)
# [] [Python 3] RGB HEX  +   | root
```

```python
# rgb 2 hex
hhex = list(map(rgb_to_hex, uniq.tolist()))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib notebook

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = uniq[:,0]
y = uniq[:,1]
z = uniq[:,2]
ax.scatter(x, y, z, color=hhex)

ax.set_xlabel('R (0~255)')
ax.set_ylabel('G (0~255)')
ax.set_zlabel('B (0~255)')
```

```python
import pandas as pd

df = pd.DataFrame(uniq, columns=['R', 'G', 'B'])

df.head()
```

### Sihouette Index

https://studying-haeung.tistory.com/10

   

- a(i) :  i  i  cluster       ( )

- b(i) :  i        ,  

$s(i) = \frac{b(i) - a(i)}{max(a(i), b(i))}$

```python
def KMeansTheBestNumberofClusters(df, cluster_min=2, cluster_max=7, random_state=27):
    from sklearn.cluster import KMeans
    from sklearn.metrics import silhouette_score
    score_dict = {}
    for k in range(cluster_min, cluster_max+1):
        kmeans = KMeans(n_clusters=k, random_state=random_state).fit(df)
        silhouette_avg = silhouette_score(df, kmeans.labels_)
        print(f'# of cluster : {k} with silhouette index {silhouette_avg}')
        score_dict[k] = silhouette_avg
    
    the_best_number_of_clusters = [k for k, v in score_dict.items() if v == max(score_dict.values())][0]
    return the_best_number_of_clusters
```

```python
KMeansTheBestNumberofClusters(df)
```

```python
kmeans = KMeans(n_clusters=4, random_state=27).fit(df)
```

```python
df['label'] = kmeans.labels_
```

```python
label_RGB = df.groupby('label').mean()
label_RGB = label_RGB.apply(round)
```

```python
hexx_list = [rgb_to_hex(label_RGB.iloc[i].tolist()) for i in label_RGB.index]
hexx_list
```

```python
for i in df['label'].unique():
    df.loc[df['label']==i, 'color'] = hexx_list[i]
```

```python
%matplotlib notebook

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = uniq[:,0]
y = uniq[:,1]
z = uniq[:,2]
ax.scatter(x, y, z, color=df['color'])

ax.set_xlabel('R (0~255)')
ax.set_ylabel('G (0~255)')
ax.set_zlabel('B (0~255)')
```
