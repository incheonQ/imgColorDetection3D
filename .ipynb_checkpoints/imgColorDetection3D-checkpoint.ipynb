```python
from PIL import Image
import numpy as np

path = "Instagram_icon.png"
img = Image.open(path)

img_arr = np.array(img)

print(img_arr.shape)
```

```python
uniq = np.unique(img_arr[:,:,0:-1].reshape(5001*5001, 3), axis=0)
```

```python
uniq.shape
```

```python
def rgb_to_hex(rgb):
    r, g, b = rgb
    r, g, b = int(r), int(g), int(b)
    return '#' + hex(r)[2:].zfill(2) + hex(g)[2:].zfill(2) + hex(b)[2:].zfill(2)
# [출처] [Python 3] RGB를 HEX로 바꾸기 + 파이썬 진법 변환|작성자 root
```

```python
hhex = list(map(rgb_to_hex, uniq.tolist()))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib notebook

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

x = uniq[:,0]
y = uniq[:,1]
z = uniq[:,2]
ax.scatter(x, y, z, color=hhex)

ax.set_xlabel('R (0~255)')
ax.set_ylabel('G (0~255)')
ax.set_zlabel('B (0~255)')
```
